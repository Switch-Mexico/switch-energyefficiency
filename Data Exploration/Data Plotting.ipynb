```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import math
import calendar
import numpy as np
import datetime
from sklearn import preprocessing
```

```python
df = pd.read_csv("PGE MedCom.csv")
temp = []
for c in df:
    if "Date" in c:
        date = c
    if ":" in c:
        temp.append(df[c].as_matrix())

data = []
for i in range(0,len(temp[0])):
    for j in range(0,len(temp)):
        data.append(temp[j][i])
```

```python
ran = 24*30*2
p = []
for i in range(0,int(len(data)/ran)):
    p.append(np.sum(data[i*ran:i*ran + ran]))
plt.plot(p)
plt.title("PGE MedInd Monthly Plot")
```

```python
ran = 24*7*2
p = []
for i in range(0,int(len(data)/ran)):
    plt.plot(data[i*ran:i*ran + ran])
plt.title("PGE MedInd Weekly Plot")
```
